```python
import pandas as pd
import sklearn
import numpy as np

import pickle
```

__Importing Telco Sample Data__

```python
telco_subset = pd.read_csv('telco_random_sample.csv');
```

```python
telco_subset.head()
```

__Data Prep__

```python
from data_cleansing import TelcoPreprocessing
X = TelcoPreprocessing(telco_subset,[],X_only=True)
```

__Importing Model__

```python
def pickle_load(name):
    PIK = str(name) + ".pickle"
    with open(PIK,"rb") as f:
        temp_item = pickle.load(f)
        return temp_item
```

```python
pipe = pickle_load('telco_pipeline')
```

__Making Predictions__

```python
from sklearn.metrics import roc_curve, roc_auc_score, log_loss
from matplotlib import pyplot as plt
```

```python
# predicted class
predictions = pipe.predict(X)

# probability of each class
probs = pipe.predict_proba(X)

# probability of the predicted class
pred_probs = (predictions == 0)*probs[:,0] + (predictions == 1)*probs[:,1]
```

```python
predictions[:5]
```

```python
probs[:5]
```

```python
pred_probs[:5]
```

```python
# create DataFrame containing Predicted Value and Confidence
d = {'Prediction (Churn = Yes)': predictions.flatten(), 'Confidence': pred_probs.flatten()};
df = pd.DataFrame(d);
df.head()
```

__Output__

```python
telco_predictions = pd.merge(telco_subset, df, how='left', left_index=True, right_index=True)
telco_predictions
```

```python
telco_predictions.to_csv('./telco_random_sample_predictions.csv', index=False)
```
