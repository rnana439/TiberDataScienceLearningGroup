Replace Blanks

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class ReplaceBlanks(TransformerMixin):    
    def transform(self, X, y=None, **kwargs):
        try:
            new = X.replace(" ", np.nan, inplace=True)
            return new
        except Exception as err:
            print('ReplaceBlanks.transform(): {}'.format(err))
        return X

    def fit(self, X, y=None, **kwargs):
        return self
```

```python
RB = ReplaceBlanks()
RB.fit_transform(X_train)
```

```python
x_cols = [c for c in X if c not in ["Churn"]]
```

```python
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        assert isinstance(X, pd.DataFrame)

        try:
            return X[self.columns]
        except KeyError:
            cols_error = list(set(self.columns) - set(X.columns))
            raise KeyError("The DataFrame does not include the columns: %s" % cols_error)
```

```python
cs = ColumnSelector(columns=x_cols)
cs.fit_transform(X_train).head()
```

```python
class TypeSelector(BaseEstimator, TransformerMixin):
    def __init__(self, dtype):
        self.dtype = dtype

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        assert isinstance(X, pd.DataFrame)
        return X.select_dtypes(include=[self.dtype])
```

```python
ts = TypeSelector("float64")
ts.fit_transform(X_train).head()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler, OneHotEncoder, Imputer
```

```python
preprocess_pipeline = make_pipeline(
    ColumnSelector(columns=x_cols),
    FeatureUnion(transformer_list=[
        ("int_features", make_pipeline(
            TypeSelector("int64"),
            Imputer(strategy="median"),
            StandardScaler()
        )),
        ("float_features", make_pipeline(
            TypeSelector("float64"),
            Imputer(strategy="median"),
            StandardScaler()
        )),
        ("categorical_features", make_pipeline(
            TypeSelector("object"),
            Imputer(strategy="most_frequent"),
            OneHotEncoder()
        ))
    ])
)
```

```python
preprocess_pipeline.fit(X_train)
```
